Los atacantes pueden descifrar contrase√±as d√©biles utilizando diversas t√©cnicas de ataque. A continuaci√≥n, te explico las m√°s comunes y c√≥mo funcionan:

---

## **1. Ataque de Fuerza Bruta**  
Este m√©todo consiste en probar todas las combinaciones posibles de caracteres hasta encontrar la contrase√±a correcta.  

üîπ **Ejemplo en Python:**  
```python
import itertools
import string

def brute_force(password, max_length=4):
    chars = string.ascii_lowercase  # Letras min√∫sculas a-z
    for length in range(1, max_length + 1):
        for attempt in itertools.product(chars, repeat=length):
            attempt = ''.join(attempt)
            if attempt == password:
                return f"Contrase√±a encontrada: {attempt}"

password = "abc"
print(brute_force(password))
```
üìå **Limitaci√≥n:** Puede tardar mucho tiempo si la contrase√±a es larga o usa caracteres especiales.  

---

## **2. Ataque por Diccionario**  
En lugar de probar todas las combinaciones, el atacante usa una lista de contrase√±as comunes para probarlas r√°pidamente.  

üîπ **Ejemplo en Python:**  
```python
def dictionary_attack(password):
    common_passwords = ["123456", "password", "qwerty", "abc123", "admin"]
    if password in common_passwords:
        return f"Contrase√±a {password} encontrada en la lista com√∫n."
    return "Contrase√±a segura."

password = "password"
print(dictionary_attack(password))
```
üìå **Limitaci√≥n:** No funciona contra contrase√±as √∫nicas o complejas.  

---

## **3. Ataque con Rainbow Tables**  
Las contrase√±as suelen almacenarse como **hashes** en las bases de datos. Los atacantes usan tablas precalculadas de hashes (Rainbow Tables) para descifrar contrase√±as r√°pidamente.  

üîπ **Ejemplo en Python:**  
```python
import hashlib

def generate_hash(password):
    return hashlib.md5(password.encode()).hexdigest()

# Simulamos una tabla de hashes
rainbow_table = {
    generate_hash("123456"): "123456",
    generate_hash("password"): "password",
    generate_hash("letmein"): "letmein",
}

hashed_password = generate_hash("password")  # Supongamos que un atacante obtiene este hash

if hashed_password in rainbow_table:
    print(f"Contrase√±a descifrada: {rainbow_table[hashed_password]}")
else:
    print("No encontrada en la Rainbow Table.")
```
üìå **Limitaci√≥n:** No funciona bien contra contrase√±as con **salt** (sal aleatoria agregada al hash).
