## **üìå 2Ô∏è‚É£ Agregar c√≥digo con problemas para analizar**

### **üìÑ C√≥digo en `src/calculadora.py`**
```python
from operaciones import sumar, dividir

def main():
    a = 10
    b = 0  # Problema: Divisi√≥n entre cero

    print(f"Suma: {sumar(a, b)}")
    print(f"Divisi√≥n: {dividir(a, b)}")  # Esto generar√° un error

if __name__ == "__main__":
    main()
```

**üõë Problema intencional**:  
- Intentamos dividir por **cero**, lo que generar√° una excepci√≥n (`ZeroDivisionError`).

---

### **üìÑ C√≥digo en `src/operaciones.py`**
```python
def sumar(a, b):
    return a + b

def dividir(a, b):
    return a / b  # No se maneja el caso de b == 0
```

**üõë Problema intencional**:  
- La funci√≥n `dividir(a, b)` **no maneja la divisi√≥n por cero**.

---

### **üìÑ C√≥digo en `tests/test_calculadora.py`**
```python
import unittest
from src.operaciones import sumar, dividir

class TestCalculadora(unittest.TestCase):
    def test_suma(self):
        self.assertEqual(sumar(5, 5), 10)

    def test_division(self):
        with self.assertRaises(ZeroDivisionError):
            dividir(10, 0)  # Esta prueba deber√≠a fallar

if __name__ == "__main__":
    unittest.main()
```
