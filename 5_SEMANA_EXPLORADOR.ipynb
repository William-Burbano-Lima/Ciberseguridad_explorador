## **OBJETIVO GENERAL**
Al finalizar la unidad, el estudiante estará en capacidad de:
- Nombrar los tipos de redes de datos por los que se conectan los dispositivos a internet.
- Definir los datos que una persona genera en internet.
- Enumerar los retos que existen al compartir información en redes de datos.
- Aplicar estrategias de protección y normativas de seguridad, incluyendo la ISO 27001.

---

## **Día 1 - Fundamentos de Redes de Datos y Componentes**

### **Objetivos del día:**
- Comprender la estructura de las redes de datos.
- Identificar los principales componentes y su funcionamiento.
- Aplicar principios de enrutamiento y resolución de nombres.

### **Contenidos:**
1. Definición de redes de datos.
2. Componentes principales:
   - Datagrama o paquete.
   - Dirección IP.
   - Servidores de nombres.
   - Enrutamiento y protocolos.
3. Introducción a la norma ISO 27001 en redes.

### **Ejemplos y Aplicaciones:**
- Simulación con Wireshark: Captura y análisis de paquetes.
- Configuración de direcciones IP en un entorno de red local.

### **Ejercicio Práctico:**
- Creación de un esquema de red con diferentes dispositivos.

### **Herramientas:**
- Wireshark.
- Packet Tracer.

---

## **Día 2 - Ataques de Red y Estrategias de Mitigación**

### **Objetivos del día:**
- Conocer los ataques de red más comunes.
- Aprender estrategias de mitigación y defensa.

### **Contenidos:**
1. Ataques de red comunes:
   - IP Spoofing.
   - SYN Flood.
   - Man in the Middle.
   - Packet Sniffing.
   - DoS y DDoS.
   - DNS Spoofing.
2. Ataques inalámbricos:
   - Wardriving.
   - Suplantación de zonas Wi-Fi.
   - Ataques a Bluetooth.
3. Normativa ISO 27001 aplicada a ataques y mitigación.

### **Ejemplos y Aplicaciones:**
- Simulación de ataques con Kali Linux.
- Implementación de medidas de seguridad en redes inalámbricas.

### **Ejercicio Práctico:**
- Uso de herramientas de escaneo de red como Nmap para detectar vulnerabilidades.

### **Herramientas:**
- Kali Linux.
- Nmap.
- Aircrack-ng.

---

## **Día 3 - Sistemas de Protección y Seguridad de Dispositivos**

### **Objetivos del día:**
- Conocer las herramientas de protección de redes.
- Aplicar controles de seguridad en dispositivos.

### **Contenidos:**
1. Sistemas de protección:
   - Firewalls.
   - Antivirus.
   - Control de acceso a la red.
   - VPNs.
2. Seguridad en dispositivos:
   - Protección de teléfonos, portátiles y tablets.
   - Seguridad en memorias USB y almacenamiento externo.
3. Relación con la norma ISO 27001.

### **Ejemplos y Aplicaciones:**
- Configuración de un firewall.
- Uso de VPNs para conexión segura.

### **Ejercicio Práctico:**
- Análisis de un caso real de brecha de seguridad y aplicación de soluciones.

### **Herramientas:**
- pfSense.
- OpenVPN.
- Windows Defender.

---

## **Día 4 - Seguridad en Aplicaciones y Configuraciones de Protección**

### **Objetivos del día:**
- Identificar vulnerabilidades en software y aplicaciones.
- Implementar medidas de protección seguras.

### **Contenidos:**
1. Seguridad en aplicaciones:
   - Software de sistema y de aplicación.
   - Orígenes no confiables.
   - Vulnerabilidades de código abierto y día cero.
   - Ataques basados en cookies y navegadores.
   - Typosquatting.
2. Estrategias de seguridad:
   - Aplicación de actualizaciones.
   - Revisar configuraciones.
   - Gestión de cookies y privacidad.
3. Aplicación de la norma ISO 27001 en software y aplicaciones.

### **Ejemplos y Aplicaciones:**
- Uso de Burp Suite para análisis de seguridad en aplicaciones web.
- Configuración segura de un navegador.

### **Ejercicio Práctico:**
- Auditoría de seguridad en una aplicación real.

### **Herramientas:**
- Burp Suite.
- OWASP ZAP.
- Firefox Security Settings.
