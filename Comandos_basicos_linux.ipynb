# **🔰 Introducción a Kali Linux: Comandos Básicos y Administración del Sistema**  
📌 **Duración**: 2 horas  
📌 **Nivel**: Principiante  
📌 **Objetivo**: Familiarizarse con los comandos esenciales de Linux en Kali para manejar archivos, directorios, permisos y usuarios.  

---

## **1️⃣ Comenzando con la Terminal en Kali Linux**  
Para abrir la terminal en Kali Linux:  
🔹 **Atajo de teclado**: `Ctrl + Alt + T`  
🔹 **Desde menú**: Aplicaciones → Terminal  

📌 **Comando básico para conocer el usuario actual:**  
```bash
whoami
```
📌 **Mostrar la versión de Kali Linux instalada:**  
```bash
lsb_release -a
```
📌 **Ver la estructura del sistema operativo:**  
```bash
uname -a
```
📌 **Ver el directorio en el que estás trabajando:**  
```bash
pwd
```
---

## **2️⃣ Navegación y Manejo de Directorios**  
📌 **Listar archivos y directorios:**  
```bash
ls
```
📌 **Listar con detalles (permisos, propietario, tamaño, fecha):**  
```bash
ls -l
```
📌 **Listar incluyendo archivos ocultos:**  
```bash
ls -la
```
📌 **Moverse entre carpetas:**  
```bash
cd /ruta/del/directorio
```
Ejemplo:  
```bash
cd /home/kali/Desktop
```
📌 **Regresar al directorio anterior:**  
```bash
cd ..
```
📌 **Ir al directorio principal del usuario:**  
```bash
cd ~
```
📌 **Crear una nueva carpeta:**  
```bash
mkdir mi_carpeta
```
📌 **Eliminar una carpeta vacía:**  
```bash
rmdir mi_carpeta
```
📌 **Eliminar una carpeta con todo su contenido:**  
```bash
rm -rf mi_carpeta
```
---

## **3️⃣ Creación, Edición y Eliminación de Archivos**  
📌 **Crear un archivo vacío:**  
```bash
touch archivo.txt
```
📌 **Ver el contenido de un archivo:**  
```bash
cat archivo.txt
```
📌 **Editar un archivo con nano (editor de texto en terminal):**  
```bash
nano archivo.txt
```
✏️ **Para guardar en Nano:**  
- `Ctrl + X` → Salir  
- `Y` → Guardar  
- `Enter` → Confirmar  

📌 **Eliminar un archivo:**  
```bash
rm archivo.txt
```
📌 **Mover o renombrar un archivo:**  
```bash
mv archivo.txt nuevo_nombre.txt
```
📌 **Copiar un archivo:**  
```bash
cp archivo.txt copia_archivo.txt
```
📌 **Buscar archivos con un nombre específico:**  
```bash
find / -name archivo.txt 2>/dev/null
```
---

## **4️⃣ Permisos y Privilegios en Archivos y Directorios**  
📌 **Ver permisos de archivos:**  
```bash
ls -l
```
Ejemplo de salida:  
```
-rwxr--r-- 1 kali kali 4096 feb 12 12:00 archivo.sh
```
📌 **Cambiar permisos (lectura, escritura, ejecución):**  
```bash
chmod 777 archivo.txt
```
📌 **Significado de valores en chmod:**  
- `7 = rwx` (lectura, escritura, ejecución)  
- `6 = rw-` (lectura, escritura)  
- `5 = r-x` (lectura, ejecución)  
- `4 = r--` (solo lectura)  

📌 **Ejemplo: dar permisos de ejecución solo al usuario:**  
```bash
chmod 700 script.sh
```
📌 **Cambiar el propietario de un archivo:**  
```bash
chown root archivo.txt
```
📌 **Cambiar grupo de un archivo:**  
```bash
chgrp root archivo.txt
```
📌 **Ejecutar un comando como superusuario (root):**  
```bash
sudo comando
```
Ejemplo:  
```bash
sudo apt update
```
📌 **Cambiar a usuario root:**  
```bash
sudo su
```
📌 **Salir del modo root:**  
```bash
exit
```
---

## **5️⃣ Gestión de Usuarios y Grupos**  
📌 **Ver usuarios en el sistema:**  
```bash
cat /etc/passwd
```
📌 **Ver grupos en el sistema:**  
```bash
cat /etc/group
```
📌 **Crear un nuevo usuario:**  
```bash
sudo useradd -m usuario
```
📌 **Asignar contraseña a un usuario:**  
```bash
sudo passwd usuario
```
📌 **Eliminar un usuario:**  
```bash
sudo userdel -r usuario
```
📌 **Agregar un usuario a un grupo (ejemplo: sudo):**  
```bash
sudo usermod -aG sudo usuario
```
📌 **Ver el grupo al que pertenece un usuario:**  
```bash
groups usuario
```
---

## **6️⃣ Comandos de Redes Básicos**  
📌 **Ver dirección IP de la máquina:**  
```bash
ip a
```
📌 **Ver conexiones activas:**  
```bash
netstat -tulnp
```
📌 **Hacer ping a otra máquina:**  
```bash
ping 192.168.1.1
```
📌 **Ver tabla de rutas:**  
```bash
route -n
```
📌 **Analizar tráfico de red con tcpdump:**  
```bash
sudo tcpdump -i eth0
```
---

## **7️⃣ Monitoreo del Sistema y Procesos**  
📌 **Ver procesos en ejecución:**  
```bash
ps aux
```
📌 **Ver procesos en tiempo real:**  
```bash
top
```
📌 **Cerrar un proceso por PID:**  
```bash
kill -9 1234
```
📌 **Ver uso del disco:**  
```bash
df -h
```
📌 **Ver uso de RAM:**  
```bash
free -m
```
