## 📌 **2. Casos Reales de Brechas de Seguridad (2 horas)**
### 🔎 **Caso 1: Sony PlayStation Network (Abril de 2011)**  
- **¿Qué pasó?**  
  - Hackers accedieron a los servidores de PlayStation Network.  
  - Se comprometieron datos de 77 millones de cuentas, incluyendo nombres, direcciones, correos y posiblemente información de tarjetas de crédito.  
- **Impacto:**  
  - Pérdida financiera de $171 millones.  
  - Se suspendió el servicio durante 23 días.  
  - Demanda colectiva y desconfianza de los usuarios.  
- **¿Cómo se pudo evitar?**  
  - Implementar autenticación multifactor (MFA).  
  - Mayor cifrado en la base de datos de usuarios.  
  - Segmentación de redes para evitar acceso masivo.  

### 🔎 **Caso 2: Dropbox (Agosto de 2012)**  
- **¿Qué pasó?**  
  - Un atacante accedió a una cuenta de empleado de Dropbox con contraseñas almacenadas en un documento de texto.  
  - Se filtraron más de 68 millones de credenciales de usuarios.  
- **Impacto:**  
  - La filtración de credenciales permitió que hackers accedieran a múltiples cuentas con contraseñas reutilizadas.  
  - Gran crisis de confianza en la empresa.  
- **¿Cómo se pudo evitar?**  
  - Uso de gestores de contraseñas en lugar de documentos de texto.  
  - Aplicar políticas de contraseñas seguras.  
  - Implementación de autenticación de dos factores (2FA).  

### 🔎 **Caso 3: Ashley Madison (2015)**  
- **¿Qué pasó?**  
  - Hackers robaron 32 millones de cuentas de una web de citas extramaritales.  
  - Se expusieron datos personales, mensajes privados y detalles financieros.  
- **Impacto:**  
  - Extorsiones y chantajes a los usuarios.  
  - Crisis de reputación y demandas contra la empresa.  
  - Suplantación de identidad y fraude con los datos filtrados.  
- **¿Cómo se pudo evitar?**  
  - Uso de cifrado en bases de datos de usuarios.  
  - Implementación de monitoreo de accesos sospechosos.  
  - Políticas estrictas de protección de datos sensibles.  

### 🔎 **Caso 4: UIDAI (Aadhaar, 2018)**  
- **¿Qué pasó?**  
  - La base de datos Aadhaar de la India fue vulnerada.  
  - Se filtraron datos personales de más de mil millones de ciudadanos.  
  - Un hacker vendía acceso a la base de datos por menos de $10.  
- **Impacto:**  
  - Riesgo de robo de identidad a gran escala.  
  - Falta de confianza en sistemas gubernamentales de identidad digital.  
- **¿Cómo se pudo evitar?**  
  - Implementar monitoreo de actividad inusual en la base de datos.  
  - Aplicar cifrado avanzado en la transmisión de datos.  
  - Uso de registros de auditoría detallados.

### 🛠️ **Paso 1: Configuración de la simulación**
1. **Elección de un entorno controlado:**  
   - Se usará un equipo con Kali Linux o herramientas en Windows como Hashcat o John the Ripper.  
   - Si no se tiene acceso a Kali Linux, se puede usar un simulador online como **CrackStation** (https://crackstation.net/) para probar contraseñas hash.  

2. **Generación de un archivo con contraseñas débiles:**  
   - Se crearán varias contraseñas débiles en un archivo `passwords.txt`, como:  
     ```
     123456
     password
     admin
     qwerty
     letmein
     ```
   - Luego, se generarán hashes de estas contraseñas usando herramientas de hashing como MD5 o SHA-256.  

---

### 🔎 **Paso 2: Intento de ataque con fuerza bruta**
📌 **Simulación en Windows con Hashcat**  
- **Comando básico:**  
  ```
  hashcat -m 0 -a 0 hashes.txt rockyou.txt --force
  ```
  (Donde `hashes.txt` contiene las contraseñas encriptadas y `rockyou.txt` es un diccionario de contraseñas reales filtradas).  

📌 **Alternativa en Kali Linux con John the Ripper**  
- **Ejecutar el ataque:**  
  ```
  john --wordlist=/usr/share/wordlists/rockyou.txt passwords.txt
  ```

📌 **Uso de CrackStation (opción sin código)**  
- Se ingresan hashes en la herramienta para ver si pueden ser descifrados fácilmente.

### 🎣 **Simulación de Phishing: "¡No caigas en la trampa!"** (30 min)  

✅ **Objetivo:** Mostrar cómo los atacantes usan el phishing para engañar a los usuarios y obtener credenciales o información sensible.  

---

### **🛠️ Paso 1: Actividad de Rompehielo - "Detecta el Phishing"** (10 min)  
📌 Se mostrarán **tres correos electrónicos** a los participantes:  
1. **Uno legítimo** (ejemplo: un correo real de un banco).  
2. **Uno sospechoso** (con errores sutiles, como un enlace engañoso o un dominio extraño).  
3. **Uno claramente fraudulento** (con faltas de ortografía y un enlace falso).  

🔹 **Preguntas para el grupo:**  
- ¿Cuál creen que es el correo falso?  
- ¿Qué señales les hicieron dudar?  

🔹 **Claves para detectar phishing:**  
✅ Revisar el **remitente** y el dominio del correo.  
✅ No hacer clic en enlaces sospechosos.  
✅ Cuidado con la urgencia ("¡Su cuenta será bloqueada en 24 horas!").  
✅ Verificar con la fuente oficial antes de dar información personal.  

---

### **🎭 Paso 2: Simulación de Phishing en Vivo** (15 min)  
🔹 **Escenario:** Se crea un **correo falso** con una URL que simula una página real (ejemplo: `banco-seguro.com` en lugar de `banco.com`).  

🔹 **Herramienta:**  
- Uso de **Zphisher** en Kali Linux para clonar un sitio web real.  
  ```
  git clone https://github.com/htr-tech/zphisher
  cd zphisher
  bash zphisher.sh
  ```
- Se elige un sitio web objetivo (ejemplo: Facebook, Gmail o un banco).  
- Se obtiene un enlace falso para que los participantes vean cómo luce una página de phishing.  

📌 **Alternativa sin código:**  
- Usar sitios web de simulación como **PhishTank** o **OpenPhish** para mostrar ejemplos reales.  

🔹 **Ejemplo de URL falsa:**  
```
https://banco-seguro.com/login (falso)
https://banco.com/login (real)
```

---

### **🚨 Paso 3: Reflexión y Recomendaciones** (5 min)  
🔹 **Preguntas clave:**  
- ¿Qué hubiera pasado si alguien ingresaba sus datos en la página falsa?  
- ¿Cómo protegerse del phishing?  

🔹 **Estrategias de protección:**  
✅ **Activar la autenticación en dos pasos (2FA).**  
✅ **Nunca hacer clic en enlaces sospechosos.**  
✅ **Verificar siempre el dominio antes de ingresar credenciales.**  
✅ **Usar extensiones como Netcraft Anti-Phishing o HTTPS Everywhere.**  
✅ **Reportar correos sospechosos en plataformas como Google Safe Browsing.**  

---

### **🚀 Conclusión de la Simulación**  
📌 **Lección clave:**

### **🕶️ Paso 2: Herramientas de Anonimato y Protección (15 min)**  

🔹 **1. Uso de VPN para ocultar la IP**  
📌 Explicación rápida: Una VPN (Red Privada Virtual) cifra el tráfico y oculta nuestra ubicación real.  

✅ **Opciones gratuitas:**  
- **ProtonVPN** (versión gratuita sin límite de datos).  
- **Windscribe** (10GB de datos al mes gratis).  

✅ **Opciones avanzadas:**  
- **Mullvad VPN** (sin necesidad de cuenta, pago anónimo).  
- **NordVPN / ExpressVPN** (servicios premium).  

📌 **Práctica:**  
1. Instalar una VPN gratuita (ejemplo: ProtonVPN).  
2. Activarla y comprobar si la IP ha cambiado en [whoer.net](https://whoer.net/).  

---

🔹 **2. Navegación anónima con Tor**  
📌 **Tor (The Onion Router)** permite navegar de forma anónima al encriptar la conexión a través de múltiples servidores.  

✅ **Pasos para usar Tor:**  
1. Descargar e instalar el [navegador Tor](https://www.torproject.org/).  
2. Abrir Tor y visitar [check.torproject.org](https://check.torproject.org/) para verificar la conexión.  

📌 **Ejemplo de uso:**  
- Acceder a páginas bloqueadas sin revelar nuestra IP.  
- Consultar información sin dejar rastros en motores de búsqueda.  

💡 **Precaución:**  
- **No iniciar sesión en cuentas personales dentro de Tor.**  
- **Evitar descargas desconocidas en la red Tor.**  

---

🔹 **3. Uso de alias y correos temporales**  
📌 **Para evitar exponer nuestro correo personal:**  
✅ [SimpleLogin](https://simplelogin.io/) – Para crear alias de correo y reenviar a nuestra cuenta real.  
✅ [Guerrilla Mail](https://www.guerrillamail.com/) – Para correos temporales sin registro.  

📌 **Práctica:**  
1. Crear un alias de correo en SimpleLogin.  
2. Usarlo para registrarse en una web sin exponer el correo real.
