### **Día 4: Cifrado Simétrico y Asimétrico**
#### **Conceptos Clave**  
- **Cifrado Simétrico:** Clave única.  
- **Cifrado Asimétrico:** Clave pública y privada.  

#### **Instalación de Python y Cryptography en Windows**  
1. Descargar Python 3.10 desde: [Python.org](https://www.python.org/downloads/).  
2. Durante la instalación: **Marcar "Add Python to PATH"** y seleccionar **Install Now**.  
3. Validar instalación:  
   - Abrir **cmd** y escribir:  
     ```bash
     python --version
     pip --version
     ```
4. Instalar la librería Cryptography:  
   ```bash
   pip install cryptography
   ```

#### **Actividad Práctica**  
- Escribir un programa en **Python** usando **Visual Studio Code**:  
   ```python
   from cryptography.fernet import Fernet

   key = Fernet.generate_key()
   cipher = Fernet(key)

   mensaje = b"Hola, ciberseguridad"
   cifrado = cipher.encrypt(mensaje)
   print("Mensaje cifrado:", cifrado)

   descifrado = cipher.decrypt(cifrado)
   print("Mensaje descifrado:", descifrado.decode())
   ```

- **Hashing:** SHA-256 para validar integridad.

#### **Actividad Práctica**  
- Programa en Python para generar un **hash SHA-256**:  
   ```python
   import hashlib

   mensaje = "Seguridad en la red"
   resultado = hashlib.sha256(mensaje.encode())
   print("Hash SHA-256:", resultado.hexdigest())
   ```

#### **Actividad Práctica**  
- Generar una clave privada y certificado:  
   ```bash
   openssl genpkey -algorithm RSA -out private.pem
   openssl req -new -key private.pem -out cert.csr
   openssl x509 -req -days 365 -in cert.csr -signkey private.pem -out cert.crt
   ```
